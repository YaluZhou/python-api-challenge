```python
# Import dependencies
import pandas as pd
import numpy as np
import requests
import json
from citipy import citipy
import os as os
import matplotlib.pyplot as plt
import csv as csv
from scipy.stats import linregress
import scipy.stats as st

# Google API Key
from config import api_key
```

```python
# Generate a random list of lats and longs
# numpy.random.uniform(low=0.0, high=1.0, size=None)
lat = np.random.uniform(low=-90.00, high=90.00, size=600)
long = np.random.uniform(low=-180.00, high=180.00, size=600)
```

```python
latlong = zip(lat, long)
```

```python
# Use citipy to generate the a list of the closest cities to your random coordinates. 
city_list = []
for c in latlong:
    city_list.append(citipy.nearest_city(c[0], c[1]))
```

```python
city_name=[]
for city in city_list:
    name = city.city_name
    city_name.append(name)
```

```python
city_name
```

```python
# Use Openweather api to get the weather data needed from those cities.
url = "http://api.openweathermap.org/data/2.5/weather?"
temps = []
humid = []
cloudness = []
wind_speed = []
lats = []
longs = []
names = []

# Build query URL
for city in city_name:
    query_url = url + "appid=" + api_key + "&q=" + city + "&units=imperial"
    response = requests.get(query_url)
    if response.status_code == 200:
        response = response.json()
        temps.append(response["main"]["temp"])
        humid.append(response["main"]["humidity"])
        cloudness.append(response["clouds"]["all"])
        wind_speed.append(response["wind"]["speed"])
        lats.append(response["coord"]["lat"])
        longs.append(response["coord"]["lon"])
        names.append(response["name"])
```

```python
city_weather_df = pd.DataFrame({"City": names,
                        "Temperature (F)": temps,
                        "Humidity (%)": humid,
                        "Cloud Coverage (%)": cloudness,
                        "Wind Speed (mph)": wind_speed,
                        "Latitude": lats, 
                        "Longitude": longs
                       })
city_weather_df.head()
```

```python
print(len(city_weather_df))
```

```python
# Scatter plot to show temperature vs latitude
temp_lat_scatter = plt.scatter(city_weather_df["Latitude"], city_weather_df["Temperature (F)"], marker="o", facecolors="red", edgecolors="black", alpha=0.75)

plt.title("Temperature (F) vs Latitude in 541 random cities", fontsize=15)
plt.xlabel("Latitude")
plt.ylabel("Temperature (F)")

plt.show()
```

```python
# Scatter plot to show humidity vs latitude
humid_lat_scatter = plt.scatter(city_weather_df["Latitude"], city_weather_df["Humidity (%)"], marker="o", facecolors="blue", edgecolors="black", alpha=0.75)

plt.title("Humidity (%) vs Latitude in 541 random cities", fontsize=15)
plt.xlabel("Latitude")
plt.ylabel("Humidity (%)")

plt.show()
```

```python
# Scatter plot to show cloudiness vs latitude
cloud_lat_scatter = plt.scatter(city_weather_df["Latitude"], city_weather_df["Cloud Coverage (%)"], marker="o", facecolors="grey", edgecolors="black", alpha=0.75)

plt.title("Cloud Coverage (%) vs Latitude in 541 random cities", fontsize=15)
plt.xlabel("Latitude")
plt.ylabel("Cloud Coverage (%)")

plt.show()
```

```python
# Scatter plot to show wind speed vs latitude
wind_lat_scatter = plt.scatter(city_weather_df["Latitude"], city_weather_df["Wind Speed (mph)"], marker="o", facecolors="green", edgecolors="black", alpha=0.75)

plt.title("Wind Speed vs Latitude in 541 random cities", fontsize=15)
plt.xlabel("Latitude")
plt.ylabel("Wind Speed (mph)")

plt.show()
```

```python
North_weather_df = city_weather_df.loc[city_weather_df["Latitude"] >= 0]
North_weather_df.head(250)
```

```python
South_weather_df = city_weather_df.loc[city_weather_df["Latitude"] < 0]
South_weather_df.head()
```

```python
# Scatter plot to show temperature and latitude in Northern Hemisphere
x_values = North_weather_df['Latitude']
y_values = North_weather_df['Temperature (F)']

# Run linear regression
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,3)) + "x + " + str(round(intercept,3))
print(f"The line equation is {line_eq}")

# Plot scatter plot
plt.scatter(x_values,y_values,marker = "o", facecolor = "red", edgecolor="black",alpha=0.75)

# Plot regression line
plt.plot(x_values,regress_values,"r-")
plt.title("Temperature vs Latitude in Northern Hemisphere", fontsize=15)
plt.xlabel("Latitude")
plt.ylabel("Temperature (F)")

print(f"R squared: {rvalue**2}")
print(f"P value:{pvalue}")
print(f"The correlation coefficient between temperature and latitude is {round(st.pearsonr(x_values,y_values)[0],2)} in Northern Hemisphere")


plt.show()
```

There is a strong negative correlation between temperature and latitude in Northern Hemisphere with statistic significance.
